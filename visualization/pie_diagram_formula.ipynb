# Core

```python
# Create an empty column to store the categories.
df['Category'] = None
```

```python
# Categorization (Adjust as needed)
for index, row in df.iterrows():
    formula = row['orig_c1']
    if pd.isna(formula):
        df.loc[index, 'Category'] = 'Other'  # Handle NaN values
    elif 'Fe' in formula and 'O' in formula: # Iron Oxides and Related
        df.loc[index, 'Category'] = 'Iron Oxides and Related'
    elif 'Ni' in formula or 'Co' in formula:
        df.loc[index, 'Category'] = 'Nickel and Cobalt Compounds'
    elif 'Mn' in formula:
        df.loc[index, 'Category'] = 'Manganese Compounds'
    elif 'Cr' in formula:
        df.loc[index, 'Category'] = 'Chromium Compounds'
    elif 'La' in formula or 'Bi' in formula or 'Nd' in formula or 'Sm' in formula: # Perovskites (broadened)
        df.loc[index, 'Category'] = 'Perovskite-like Compounds'
    elif 'Zn' in formula or 'Mg' in formula or 'Al' in formula:
        df.loc[index, 'Category'] = 'Zinc/Magnesium/Aluminum Compounds'
    elif any(metal in formula for metal in ['Cu', 'Ag', 'Au','Mo', 'Si', 'Cr', 'Fe','Mn','Ni', 'Co', 'Pd', 'Pt', 'Ga', 'Ge', 'Ce', 'Pr', 'Ti', 'Tb', 'Ca', 'Eu']):
        df.loc[index, 'Category'] = 'Metals and Alloys'
    elif 'C' in formula:
         df.loc[index, 'Category'] = 'Carbon-containing Compounds'
    elif 'Ti' in formula:
        df.loc[index, 'Category'] = 'Titanium Compounds'
    elif 'Ba' in formula or 'Sr' in formula:
        df.loc[index, 'Category'] = 'Barium/Strontium Compounds'
    elif 'Se' in formula:
        df.loc[index, 'Category'] = 'Selenium compounds'
```

```python
# Create the dictionary
formula_dict = df.groupby('Category')['orig_c1'].apply(list).to_dict()

# Print the dictionary
print("Category Dictionary:")
print(formula_dict)
```

```python
# Print the categories
categories = df['Category'].unique()
print("\nAvailable Categories:")
print(categories)
```

```python
# Pie Chart with Legend and percentages on chart, and improved label placement ---
category_counts = df['Category'].value_counts()
labels = category_counts.index
sizes = category_counts.values
percentages = 100. * sizes / sizes.sum()

# Labels for legend WITH percentages
legend_labels = ['{0} - {1:1.1f} %'.format(i, j) for i, j in zip(labels, percentages)]

fig, ax = plt.subplots(figsize=(10, 4.5))  # Increased figure size for even more space

# Pie chart with percentages on the slices
wedges, texts, autotexts = ax.pie(sizes,
       autopct='%1.1f%%',  # Add percentage to pie slices
       labels=None, # remove labels from slices
       startangle=290,
       pctdistance=0.65,  # Adjust the distance of percentage labels from the center
       textprops=dict(color="black", fontsize=16)) # makes the numbers on the slices white

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Add a legend
plt.legend(legend_labels, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=12) # Reduced fontsize to make the legend fit.

plt.title('Distribution of formula core', fontsize=25, ha='center')
plt.tight_layout(pad=0.5)   # Adjust layout to make room for the legend

threshold = 3  # Trashold
radius = 1.52  # Radius

# Signature coordinates
annotations = []
for i, (wedge, pct) in enumerate(zip(wedges, percentages)):
    angle = (wedge.theta2 - wedge.theta1)/2 + wedge.theta1
    y = np.sin(np.deg2rad(angle))
    x = np.cos(np.deg2rad(angle))
    horizontalalignment = "right" if x < 0 else "left"

    if pct >= threshold:
        autotexts[i].set_text(f'{pct:.1f}%')
    else:
        autotexts[i].set_text("")
        xytext = (radius * x, radius * y)

        # Check the overlap and adjust position
        for prev in annotations:
            if np.linalg.norm(np.array(prev) - np.array(xytext)) < 0.2:
                xytext = (xytext[0], xytext[1] + 0.15)

        ax.annotate(legend_labels[i], xy=(x, y), xytext=xytext,
                    horizontalalignment=horizontalalignment,
                    arrowprops=dict(arrowstyle="-", connectionstyle=f"angle,angleA=0,angleB={angle}"),
                    fontsize=16)
        annotations.append(xytext)

plt.show()
```

# Shell

```python
# Create an empty column to store the categories.
df['Category'] = None
```

```python
# Categorization (Adjust as needed)
for index, row in df.iterrows():
    formula = row['orig_c2']
    if pd.isna(formula):
        df.loc[index, 'Category'] = 'Other'  # Handle NaN values
    elif formula == '0':  # Handle '0' as an outlier
        df.loc[index, 'Category'] = 'No shell'
    elif 'Fe' in formula and 'O' in formula:
        # Iron Oxides and Related
        if 'Co' in formula and 'Fe' in formula and 'O' in formula and ((not 'Mn' in formula) or ('Co' in formula and 'Mn' in formula and 'Co' not in formula.split('Fe')[0])):
            df.loc[index, 'Category'] = 'Cobalt Ferrites (CoFe2O4 and related)'
        elif 'Ni' in formula and 'Fe' in formula and 'O' in formula:
            df.loc[index, 'Category'] = 'Nickel Ferrites (NiFe2O4 and related)'
        else:
            df.loc[index, 'Category'] = 'Iron Oxides and Related'
    elif 'Ni' in formula or 'Co' in formula:
        df.loc[index, 'Category'] = 'Nickel and Cobalt Compounds'
    elif 'Mn' in formula:
        df.loc[index, 'Category'] = 'Manganese Compounds'
    elif 'Cr' in formula:
        df.loc[index, 'Category'] = 'Chromium Compounds'
    elif 'La' in formula or 'Bi' in formula or 'Nd' in formula or 'Sm' in formula: # Perovskites (broadened)
        df.loc[index, 'Category'] = 'Perovskite-like Compounds'
    elif 'Zn' in formula or 'Mg' in formula or 'Al' in formula:
        df.loc[index, 'Category'] = 'Zinc/Magnesium/Aluminum Compounds'
    elif any(metal in formula for metal in ['Cu', 'Ag', 'Au','Mo', 'Si', 'Cr', 'Fe','Mn','Ni', 'Co', 'Pd', 'Pt', 'Ga', 'Ge', 'Ce', 'Pr', 'Ti', 'Tb', 'Ca', 'Eu']):
        df.loc[index, 'Category'] = 'Metals and Alloys'
    elif 'Te' in formula:
        df.loc[index, 'Category'] = 'Telluride Compounds'
    elif 'Ti' in formula:
        df.loc[index, 'Category'] = 'Titanium Compounds'
    elif 'Gd' in formula:
        df.loc[index, 'Category'] = 'Gadolinium Compounds'
    elif 'Ba' in formula or 'Sr' in formula:
        df.loc[index, 'Category'] = 'Barium/Strontium Compounds'
    elif 'Se' in formula:
        df.loc[index, 'Category'] = 'Selenium compounds'
    elif 'Zn' in formula:
        df.loc[index, 'Category'] = 'Zinc compounds'
    elif 'Si' in formula:
        df.loc[index, 'Category'] = 'Silicates/Silicon compounds'
    else:
        df.loc[index, 'Category'] = 'Mixed Compounds'
```

```python
# Pie Chart with Legend and percentages on chart, and improved label placement ---
category_counts = df['Category'].value_counts()
labels = category_counts.index
sizes = category_counts.values
percentages = 100. * sizes / sizes.sum()

# Labels for legend WITH percentages
legend_labels = ['{0} - {1:1.1f} %'.format(i, j) for i, j in zip(labels, percentages)]

fig, ax = plt.subplots(figsize=(10, 4.5))  # Increased figure size for even more space

# Pie chart with percentages on the slices
wedges, texts, autotexts = ax.pie(sizes,
       autopct='%1.1f%%',  # Add percentage to pie slices
       labels=None, # remove labels from slices
       startangle=280,
       pctdistance=0.65,  # Adjust the distance of percentage labels from the center
       textprops=dict(color="black", fontsize=16)) # makes the numbers on the slices white

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Add a legend
plt.legend(legend_labels, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=14) # Reduced fontsize to make the legend fit.

plt.title('Distribution of formula shell', fontsize=25, ha='center')
plt.tight_layout(pad=0.5)   # Adjust layout to make room for the legend

# Threshold for placing label outside the pie
threshold = 3
fixed_label_y = -1.1   # Fixed y-coordinate for all labels (below the pie)
fixed_left_x = -0.4    # Fixed x-coordinate for the left side
fixed_right_x = 0.4    # Fixed x-coordinate for the right side
y_spacing = 0.13       # Minimum vertical spacing between labels

# Lists to store already placed y-coordinates for the left and right sides
left_y_coords = []
right_y_coords = []

for i, (wedge, pct) in enumerate(zip(wedges, percentages)):
    # Calculate the central angle of the wedge
    angle = (wedge.theta2 - wedge.theta1) / 2 + wedge.theta1
    # Calculate the coordinates on the circumference of the pie (starting point of the arrow)
    x = np.cos(np.deg2rad(angle))
    y = np.sin(np.deg2rad(angle))

    if pct >= threshold:
        autotexts[i].set_text(f'{pct:.1f}%')  # Label inside the wedge
    else:
        autotexts[i].set_text("")
        # Determine which side the label should be on and fix the x-coordinate,
        # and set y to the fixed_label_y (with slight adjustment if overlapping)
        if x < 0:
            label_x = fixed_left_x
            ha = "right"
            side_y_coords = left_y_coords
        else:
            label_x = fixed_right_x
            ha = "left"
            side_y_coords = right_y_coords

        # Start with the base fixed y and, if there is overlap, shift downward
        label_y = fixed_label_y
        while any(abs(label_y - prev) < y_spacing for prev in side_y_coords):
            label_y -= y_spacing
        side_y_coords.append(label_y)

        xytext = (label_x, label_y)

        # Use connectionstyle "angle,angleA=0,angleB=90,rad=0" for a sharp-angled arrow
        ax.annotate(legend_labels[i],
                    xy=(x, y),      # Point on the circumference of the pie (arrow starting point)
                    xytext=xytext,  # Fixed label position (below the pie)
                    horizontalalignment=ha,
                    fontsize=14,
                    arrowprops=dict(
                        arrowstyle="-",
                        connectionstyle="angle,angleA=0,angleB=90,rad=0",  # Arrow with a sharp angle
                        lw=1.2)
                   )

plt.show()
```
