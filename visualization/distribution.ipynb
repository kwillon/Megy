```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Megy_final.csv')
```

```python
list_to_remove = [
    "Unnamed: 0",
    "id",
    "x",
    "y",
    "z"]
df = df.drop(columns=list_to_remove, errors='ignore')
```

```python
# Set global font size for all plots to 16
plt.rcParams.update({'font.size': 16})
# Optionally, set a seaborn style without gridlines
sns.set_style("white")
```

```python
# Get list of numeric columns to display distributions
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print("Numeric columns:", numeric_cols)
```

```python
# Plot histograms for each numeric column with KDE
for col in numeric_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True, color="skyblue")
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(False)  # Remove grid
    plt.tight_layout()
    plt.show()
```

```python
# Create a horizontal boxplot for all numeric columns
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[numeric_cols], orient='h', palette="Set2")
plt.title("Boxplot of Numeric Columns")
plt.xlabel("Value")
plt.ylabel("Variable")
plt.grid(False)  # Remove grid
plt.tight_layout()
plt.show()
```

```python
# Create a pairplot (scatter plot matrix) for a subset of numeric columns
subset_cols = numeric_cols[:10]
pair_plot = sns.pairplot(df[subset_cols], diag_kind="kde", height=3)
pair_plot.fig.suptitle("Pairplot of Selected Numeric Columns", y=1.02, fontsize=16)
plt.show()
```
