Используем фунцию Composition  для каждого соеденения и сохраним, как список, в котором уже будем считать средную электроотрицательность по формуле

````python
list1 = []
for formula in df['Core composition (formula)']:
  comp = Composition(formula)
  list1.append(comp.average_electroneg)

df['Electronegativity core'] = list1

df['Electronegativity core'].values
````
